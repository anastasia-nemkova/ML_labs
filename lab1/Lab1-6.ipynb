2. Сколько процентов студентов посещают внеклассовые активности?

3. Правда ли, что в среднем девушки сдают экзамены лучше, чем мужчины

````python
df["Total_Score"] = df[["Previous_Scores", "Exam_Score"]].sum(axis=1) # создадим столбец с общим баллом за два экзамена

if df[df["Gender"] == "Male"]["Total_Score"].mean() < df[df["Gender"] == "Female"]["Total_Score"].mean():
    print("Yes")
elif df[df["Gender"] == "Male"]["Total_Score"].mean() > df[df["Gender"] == "Female"]["Total_Score"].mean():
    print("No")
else:
    print("Same")
````

4. Выведите 25 и 75 перцентили времени сна студентов

````python
df["Sleep_Hours"].quantile(q=[0.25, 0.75])
````

5. Правда ли, что те, кто спит больше(75 перцентиль и выше) сдает экзамен лучше, чем студенты с недосыпом(25 перцентиль и меньше)

````python
per25 = np.percentile(df["Sleep_Hours"], 25)
per75 = np.percentile(df["Sleep_Hours"], 75)
sleep_more = df[df["Sleep_Hours"] >= per75]
sleep_litlle = df[df["Sleep_Hours"] <= per25]

if sleep_more["Total_Score"].mean() > sleep_litlle["Total_Score"].mean():
    print("Yes")
elif sleep_more["Total_Score"].mean() < sleep_litlle["Total_Score"].mean():
    print("No")
else:
    print("Same")

````

6. Влияет ли физическая активность на результаты экзамена?

````python
correlation = df["Physical_Activity"].corr(df["Total_Score"])

if -0.1 <= correlation <= 0.1:
    print("Физическая активность не оказывает значительного влияния")
elif correlation > 0.1:
    print("Физическая активность влияет положительно")
else:
    print("Физическая активность влияет отрицательно")
````

7. Правда ли, что слуденты, чьи родители имеют высшее образование(College || Postgraduate), сдают экзамены лучше, чем остальные?

````python
high_ed = df[(df["Parental_Education_Level"] == "College") | (df["Parental_Education_Level"] == "Postgraduate")]["Total_Score"].mean()
no_high_ed = df[df["Parental_Education_Level"] == "High School"]["Total_Score"].mean()

if high_ed > no_high_ed:
    print("Yes")
elif high_ed < no_high_ed:
    print("No")
else:
    print("Same")
````

8. Как влияет участие родителей на результаты экзаменов

````python
d = {"High": 2, "Medium": 1, "Low": 0}
df["Parental_Involvement"] = df["Parental_Involvement"].map(d)

correlation = df["Parental_Involvement"].corr(df["Total_Score"])
if -0.1 <= correlation <= 0.1:
    print("Участие родителей не оказывает значительного влияния")
elif correlation > 0.1:
    print("Участие родителей влияет положительно")
else:
    print("Участие родителей влияет отрицательно") 
````

9. Правда ли, что в семьях с доходом больше среднего дети сдают экзамены лучше?

````python
high_income = df[df["Family_Income"] == "High"]
no_high_income = df[df["Family_Income"] != "High"]

result_high = high_income["Total_Score"].mean()
result_no_high = no_high_income["Total_Score"].mean()

if result_high > result_no_high:
    print("Да")
else:
    print("Нет")
````

10. Посчитайте долю студентов сдавших экзамен на более чем 70 баллов в семьях с низким доходом и отсутствием доступа к интернету

````python
low_income_without_internet = (df["Family_Income"] == "Low") & (df["Internet_Access"] == "No")
high_result = (df["Exam_Score"] > 70) & (df["Previous_Scores"] > 70)
good = df[(low_income_without_internet & high_result)]

all = df[low_income_without_internet].shape[0]
more = good.shape[0]
perc = more / all * 100
print(perc)
````

4. Постройте box plot отображающий распределение количества сна детей сдавших экзамены на оценку ниже и выше 85 баллов

````python
def category(row):
    if row["Previous_Scores"] >= 85 and row["Exam_Score"] >= 85:
        return "More_85"
    else:
        return "Less_85"

df["Result_Category"] = df.apply(category, axis=1)

sns.boxplot(x="Result_Category", y="Sleep_Hours", data=df)
plt.title("Распределение количества часов сна")
plt.xlabel("Оценка")
plt.ylabel("Количество часов сна")
````

7. Постройте HeatMap где ось Х - процент посещаемость, Y - количество занятий с репетитором, а Z - балл за итоговый экзамен*

\* Задание на доп. балл

````python
plt.figure(figsize=(20, 8))
heatmap_data = df.pivot_table(index="Tutoring_Sessions", columns="Attendance", values="Exam_Score")

min_value = df.loc[df["Exam_Score"] > 0, "Exam_Score"].min()
max_value = df["Exam_Score"].max()

sns.heatmap(heatmap_data, linewidths=.5, annot=True, fmt=".0f", vmin=min_value, vmax=max_value, cbar_kws={"label": "Exam_Score"})
plt.title("Зависимость результатов экзамена от процента посещаемости и количества занятий с репетитором", pad=20)
plt.show()
````
